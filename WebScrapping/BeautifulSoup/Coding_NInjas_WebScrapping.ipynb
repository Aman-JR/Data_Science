- Print the title of all 20 books which are present on first page https://books.toscrape.com/ of this website.

```python
response = requests.get('https://books.toscrape.com/')
text_data = response.text
data = BeautifulSoup(text_data,'html.parser')
li_Book_class = data.find_all(class_ = "product_pod")
for i in li_Book_class:
    print(i.h3.a['title'])
```

- Print the name of all categories which are present this website.

```python
response = requests.get('https://books.toscrape.com/')
text_data = response.text
data = BeautifulSoup(text_data,'html.parser')
Book_Category_class = data.find(class_ ="nav nav-list")
ans = Book_Category_class.li.ul
Book_Names =  ans.find_all('li')
for i in Book_Names:
    print(i.text.strip())
```

#### BOOK DETAILS
- Create the book details Book_Name, Link, Price, Quantity_In_Stock

```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup

All_Urls = ["https://books.toscrape.com/catalogue/page-1.html"]
current_page = "https://books.toscrape.com/catalogue/page-1.html"
base_url = "https://books.toscrape.com/catalogue/"

response = requests.get(current_page)
Books = []
while(response.status_code == 200):
    data = BeautifulSoup(response.text, 'html.parser')
    next_page = data.find(class_ = 'next')
    if next_page == None:
        break
    next = next_page.a['href']
    next_page_url = base_url + next

    All_Urls.append(next_page_url)
    response = requests.get(next_page_url)

Books = []
for i in range(0,len(All_Urls)):
    response = requests.get(All_Urls[i])
    data = BeautifulSoup(response.text,'html.parser')
    books_url = data.find_all(class_ = 'product_pod')
    for j in books_url:
        base_url = 'https://books.toscrape.com/catalogue/'
        link = j.h3.a['href']
        # print(base_url+link)
        Books.append(base_url+link)

dataF = []
for i in range(0,len(Books)):
    response = requests.get(Books[i])
    data = BeautifulSoup(response.text, 'html.parser')
    title = data.h1.string
    Link = Books[0]
    Price = data.find(class_ = "price_color").text
    Price = float(re.search('[\d.]+',Price).group())
    Quantity_In_Stock = data.find(class_ = 'instock availability').text
    Quantity_In_Stock = int(re.search('\d+',Quantity_In_Stock).group())
    print(title,Link,Price,Quantity_In_Stock)
    dataF.append([title,Link,Price,Quantity_In_Stock])

df = pd.DataFrame(dataF)
df['Price'] = df['Price'].astype(str)
```

```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup


All_Urls = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html']


Books = []
for i in range(0,len(All_Urls)):
    response = requests.get(All_Urls[i])
    data = BeautifulSoup(response.text,'html.parser')
    books_url = data.find_all(class_ = 'product_pod')
    for j in books_url:
        base_url = 'https://books.toscrape.com/catalogue/'
        link = j.h3.a['href']
        # print(base_url+link)
        Books.append(base_url+link)

dataF = []
for i in range(0,len(Books)):
    response = requests.get(Books[i])
    data = BeautifulSoup(response.text, 'html.parser')
    title = data.h1.string
    Link = Books[0]
    Price = data.find(class_ = "price_color").text
    Price = float(re.search('[\d.]+',Price).group())
    Quantity_In_Stock = data.find(class_ = 'instock availability').text
    Quantity_In_Stock = int(re.search('\d+',Quantity_In_Stock).group())
    print(title,Link,Price,Quantity_In_Stock)
    dataF.append([title,Link,Price,Quantity_In_Stock])

df = pd.DataFrame(dataF, columns= ['Title', 'Link', 'Price', 'Quantity in Stock'])
```
