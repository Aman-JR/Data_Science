```python
"""
1. Get current height
2. Scroll to that height
3. Wait for some time
4. If current height == Previous Height Break
5. Scroll again to new current Height
"""
```

```python
# Get current height
current_height = driver.execute_script("return document.body.scrollHeight")

while True:
    (
        driver.
        execute_script(f"window.scrollTo(0,{current_height})")
    )
    time.sleep(4)
    new_hight = driver.execute_script("return document.body.scrollHeight")
    if new_hight == current_height:
        break
    else:
        current_height = new_hight
```
