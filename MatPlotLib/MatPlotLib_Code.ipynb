#### Color, Marker, LineWidth, Labels for axis, Datasets and Title, Axis, Grid, Text

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,5,0.1)
y2 = x **2
y3 = x **3

plt.grid()
plt.plot(x,y2,color = "Red",marker = 'o',label = "x^2")
plt.plot(x,y3,'--',color = "Green", linewidth = 3, label = "x^3")

# To print the labels
plt.legend()

# To set the title of the graph
plt.title("Learn The Basics")

# To plot the axis according to the need axis([x_start,x_end,y_start,y_end])
# plt.axis([0,10,0,150])

# To label the x axis
plt.xlabel("X")

# TO label the y axis
plt.ylabel("pow(y,x)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0,1,2,3,4,5,6,7,9,10])
y1 = np.array([1,2,5,10,17,26,37,50,82,101])

x2 = np.arange(0,40,2)
y2 = 2*x2

# plt.axis([0,40,0,100])
print('0 to 10')
print('25 to 35')
plt.grid()
plt.plot(x1,y1,'--',color = 'Red', label = 'X,Y = X^2 + 1')
plt.plot(x2,y2,'--',color = 'Green',label = 'X,y = 2*X')
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
