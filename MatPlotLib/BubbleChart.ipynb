#### BUBBLE CHART

```python
import matplotlib.pyplot as plt

Year = [2012, 2013, 2014, 2015, 2016, 2017]
Salary = [12, 13, 14, 17, 19, 20]
Population = [100, 120, 180, 250, 300, 370]
```

```python
# To plt the bubble Chart use the scatter plt and the assign size.
# size can bw assign as s = []
plt.scatter(Year,Salary,s = Population, color = 'Red')
plt.title("Bubble Chart")
plt.show()
```

```python
# Using Alpha in the chart (0,1) anything between 0 to 1 
plt.scatter(Year,Salary,s = Population, color = 'Red',alpha= 0.5)
plt.title("Bubble Chart")
plt.show()
```

```python
import numpy as np
c = np.arange(len(Year))
c
```

```python
# Giving the array for the colors
# Edge color
plt.scatter(Year,Salary,s = Population, c = c,alpha= 0.5, edgecolor = 'Black', marker = "p")
plt.title("Bubble Chart")
plt.show()
```

```python
# To add text in the chart
plt.scatter(Year,Salary,s = Population, c = c,alpha= 0.5, edgecolor = 'Black', marker = "p")
plt.title("Bubble Chart")
for i in range(0,len(Year)):
    plt.text(Year[i]-0.1,Salary[i]+0.4 , Population[i])
plt.show()
```

#### We are given data of Microsoft Corporations, it contains the gross annual revenue in billion U.S dollar and the number of employees in thousands and year. 
##### Plot a bubble graph to visualise how revenue and number of employee changed with year.

##### Find years where there is a drastic increase in gross revenue of Microsoft Corporations (from previous and next year both). Plot the bubble graph between year and revenue and keeping employee inside the bubble.

#### Print the year, revenue and number of the employee where there is a drastic increase in revenue ( Top 2 ).

#### Note:For finding the year where there is a drastic increase in gross revenue check whether the revenue of that year is greater than in previous years and greater than next year.
#### Here years should be printed in ascending order.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
employees = np.array([61,71,79,91,93,89,90,94,99,128,118,114,124,131])
year = np.array([2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018])
revenue = np.array([39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36])
```

```python
c = np.arange(len(year))
c
```

```python
plt.grid()
plt.scatter(year,revenue,s = employees,c = c,alpha= 0.5,edgecolor = 'Black')
for i in range(0,len(year)):
    plt.text(year[i]-0.6,revenue[i]+1.9,employees[i])
plt.title("Employees Chart")
plt.show()
```
