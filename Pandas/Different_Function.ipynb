##### Find Frequency of the casualty in Red Corridor States and in Jammu and Kashmir? 
##### Frequency = Total Casualty/ Total number of year
##### Note : Red Corridor state includes Jharkhand, Odisha, Andhra Pradesh, and Chhattisgarh
##### Casualty = Killed + Wounded
##### Answer = Freq1 , Freq2


```python
import numpy as np

# Creating new col = col[Killed] + col[wounded]
df['Casualty'] = df['Killed'] + df['Wounded']

# Replace NaN value with 0
df.fillna(0,inplace=True)
print(df.head(10))

# Create the set of the column name and count the length for total Number of years
T_year = len(set(df["Year"]))

new_df = df[df['Country'] == 'India']
# new_df['Casualty'] = new_df['Killed'] + new_df['Wounded']
# print(new_df['City'].value_counts())

# Creating Red Corridor states data
Red_C = new_df[(df.State == 'Jharkhand') | (df.State == 'Odisha') | (df.State == 'Andhra Pradesh') | (df.State == 'Chhattisgarh')]

# print(new_df.head())
# Creating J & K data
JK = new_df[df.State == 'Jammu and Kashmir']

sumJk = int(np.sum(JK['Casualty']))
sumRed_C = int(np.sum(Red_C['Casualty']))

Freq1 = sumRed_C//T_year
Freq2 = sumJk//T_year
print(Freq1,Freq2)
```

```python
# Find the Max value from the Killed Column
Max_Killed = max(df['Killed'])


# With the help of max_killed value find the Group who did it.
# df['Where killed = MaxValue'][Group]
group = df[df['Killed'] == Max_Killed]['Group'].values[0]

# With the same logic find the Country where the most number of people killed
Country = df[df['Killed'] == Max_Killed]['Country'].values[0]


print(int(Max_Killed),Country,group)
```
